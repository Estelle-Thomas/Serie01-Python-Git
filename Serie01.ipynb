## Minimal script to see the ECG Signal 

```python
# load library
import numpy as np
import matplotlib.pyplot as plt

# load data 
ecg_raw = np.loadtxt('data/ECGu.txt')
print(ecg_raw.shape)
print(ecg_raw)

# plot data
plt.plot(ecg_raw)
plt.title('ECG Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

```python
# Create t the time vector in seconds
sampling_freqeuncy = 1000  # Sampling frequency in Hz
T = 1 / sampling_freqeuncy
N = len(ecg_raw)
start = 0
stop = (N-1) * T 
t = np.linspace(start, stop, N) # Time vector

plt.plot(t, ecg_raw)
plt.title('ECG Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()

```

```python
# Rescale the ecg signal to knowing that 1 unit = 1024 microV
ecg = ecg_raw * 1024e-6  # Convert to mV , A/D converter gain : 1024 µV per unit

# plot the ecg signal in mV
plt.plot(t, ecg)
plt.title('ECG Signal')
plt.xlabel('Time (seconds)')
plt.ylabel('Amplitude (mV)')
plt.grid()
plt.show()
```
