## Minimal script to see the ECG Signal 

```python
# load library
import numpy as np
import matplotlib.pyplot as plt

# load data 
ecg_raw = np.loadtxt('data/ECGu.txt')
print(ecg_raw.shape)
print(ecg_raw)

# plot data
plt.plot(ecg_raw)
plt.title('ECG Signal')
plt.xlabel('Time (samples)')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```
